```python
import pandas as pd
```

```python
df = pd.read_csv('short_answers_merged.csv', encoding='unicode_escape')
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from pprint import pprint
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df['student_answers_embeddings'] = df['all_answers'].apply(lambda x : model.encode([x]))
```

```python
df['professor_answers_embeddings'] = df['answers'].apply(lambda x : model.encode([x]))
```

```python
df.shape
```

```python
df['cosine_similarity_score'] = df.apply(lambda row: cosine_similarity(row['student_answers_embeddings'],row['professor_answers_embeddings']),axis=1)
```
